# Q-Q Plot_12.9

```
#Q-Q plot

import numpy as np 
import pylab 
import scipy.stats as stats
# generating random varibles that having N(0,1)
std_normal = np.random.normal(loc = 0, scale = 1, size=1000)

# 0 to 100th percentiles of std-normal making.
for i in range(0,101):
    print(i, np.percentile(std_normal,i))
```

```
import matplotlib.pyplot as plt
```

```
# generate 100 samples from N(20,5)
measurements = np.random.normal(loc = 25, scale = 5, size=1000) 
#try size=10000

#plt.xlim(-1,1)
# to check given data follows Gaussion or not, we can see with scipy.stats library using graph.
# how Q-Q plot works: (On Y-axis): take data  -> sort them -> measure percentiles (0-100).   (on X-axis): take random data which follows N(0,1) -> sort them -> measure percentiles(0-100). 
#                                  If you plot those persentiles you will get +ve straight line then it means the data follows guassion distribution.
#                   -> Disadvantage: if your are taking small data then it will not shows you best line results.(some of the points are present away from the line)
stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show() 
```

```
# generate 100 sanples from N(20,5)
measurements = np.random.normal(loc=0, scale = 1, size=10000) 
#try size=1000
#plt.xlim(-4, 4)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```

```
# generate 1000 sanples from uniform distibution and we check how the Q-Q plots are responding with non-gaussion distibution.
measurements = np.random.uniform(low=0, high=1, size=1000)

stats.probplot(measurements, dist="norm", plot=pylab)
pylab.show()
```
